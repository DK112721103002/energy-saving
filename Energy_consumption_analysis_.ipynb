```python
import pandas as pd
import numpy as np
import datetime

np.random.seed(0)  # For reproducibility

start_date = datetime.date(2020, 1, 1)
end_date = datetime.date(2020, 12, 31)
date_range = pd.date_range(start=start_date, end=end_date, freq='D')

energy_consumption = np.random.normal(loc=50, scale=10, size=len(date_range))

# Create a DataFrame
energy_data = pd.DataFrame({
    'Date': date_range,
    'Energy_Consumption_kWh': energy_consumption
})

# Display the first few rows of the dataset
print("Sample energy consumption data:")
print(energy_data.head())

# Summary statistics of the dataset
print("\nSummary statistics of the dataset:")
print(energy_data.describe())

# Data visualization
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(energy_data['Date'], energy_data['Energy_Consumption_kWh'], marker='o', linestyle='-')
plt.title('Energy Consumption Over Time')
plt.xlabel('Date')
plt.ylabel('Energy Consumption (kWh)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
